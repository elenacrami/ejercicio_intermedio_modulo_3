3. Preparación de los datos

```python
# homogeneizo los nombres de las columnas para que todas empiecen por minúsculas
# 1.creo un dict comprehension para iterar por todas las columnas de DataFrame
columnas_minusculas = {columna: columna.lower() for columna in df_spotify.columns}

# 2.compruebo que se han realizado los cambios correctamente
columnas_minusculas
```

```python
# 3.aplico el método .rename()
df_spotify.rename(columns = columnas_minusculas, inplace = True)

# 2.verifico los cambios en el DataFrame
df_spotify.head()
```

```python
# Creo una función de limpieza para cambiar el tipo de valor en las columnas:
# danceability, energy, key, loudness, speechiness, acousticness, instrumentalness

def cambio_a_float(cadena):

    try:
        # reemplaz las comas por puntos en la cadena
        return float(cadena.replace(",", "."))
    
    except:
        return np.nan
```

```python
columnas_a_cambiar = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]
for col in columnas_a_cambiar:
    df_spotify[col] = df_spotify[col]. apply(cambio_a_float)
```

```python
df_spotify.head(3)
```

```python
# Compruebo que los datos son del tipo correcto, es decir, float
df_spotify.info()

# También se podría comprobar con df_spotify.dtypes
```

```python
# Creo una función de limpieza para homogeneizar los valores de las columnas track, title y album a minúscula.

cols_a_cambiar = ["track", "title", "album"]

for col in cols_a_cambiar:
    if df_spotify[col].dtype == "object": # agrego la condición de que si el tipo de valor dentro de cualquiera de las columnas, es de tipo "object", me aplique el cambio
       df_spotify[col] = df_spotify[col].str.lower()

    else: # de no ser así, le indico que lo deje como está
        pass
```

```python
df_spotify.sample(3)
```

```python
# Creo una función de limpieza para reemplazar los símbolos $ por "s".
df_spotify["artist"] = df_spotify["artist"].str.replace("$", "s")
```

```python
df_spotify.sample(15)
```

```python
# Calculo la media y la desviación estándar de "danceability", "acousticness" y "speechiness" de cada artista. 
# Ordeno los resultados de mayor a menor en base a la media.

media_desv = df_spotify.groupby("artist")[["danceability", "acousticness", "speechiness"]].agg(["mean", "std"])
```

```python
# Muestro los resultados en tres dataframes diferentes, uno para cada métrica ("danceability", "acousticness" y "speechiness").
# DataFrame para danceability
df_danceability = media_desv[['danceability']]
df_danceability.columns = ['danceability_mean', 'danceability_std']

# DataFrame para acousticness
df_acousticness = media_desv[['acousticness']]
df_acousticness.columns = ['acousticness_mean', 'acousticness_std']

# DataFrame para speechiness
df_speechiness = media_desv[['speechiness']]
df_speechiness.columns = ['speechiness_mean', 'speechiness_std']
```

```python
# Ordeno los resultados de mayor a menor en base a la media.
df_danceability = df_danceability.sort_values(by='danceability_mean', ascending=False)
df_acousticness = df_acousticness.sort_values(by='acousticness_mean', ascending=False)
df_speechiness = df_speechiness.sort_values(by='speechiness_mean', ascending=False)
```

```python
# Agrupo por artista y calculo la media de visitas y de likes.
media_visitas_likes = df_spotify.groupby("artist")[["views", "likes"]].mean()
media_visitas_likes.sample(10)
```

```python
# Selecciono los 10 artistas que más visitas han tenido
top_10_artistas = media_visitas_likes.sort_values(by='views', ascending=False).head(10)
top_10_artistas
```

```python
# Selecciono los 10 artistas que más likes han tenido
top_10_artistas2 = media_visitas_likes.sort_values(by='likes', ascending=False).head(10)
top_10_artistas2
```
